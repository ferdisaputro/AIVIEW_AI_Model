```python
from deepface import DeepFace
import os
import cv2


model = DeepFace.build_model("VGG-Face")

# result = DeepFace.represent(image, model_name="VGG-Face", enforce_detection=False)
```

```python
root = 'datasets/ImageData/trainingData/'
output_dir = 'temp_result'

# Create output directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

for folder in os.listdir(root)[:1]:  # Skip the first entry if it's not a folder
	folder_path = os.path.join(root, folder)
	print(f"Processing folder: {folder_path}")
	if os.path.isdir(folder_path):
		for file in os.listdir(folder_path):
			image_path = os.path.join(folder_path, file)
			
			# Read the image
			img = cv2.imread(image_path)
			
			result = DeepFace.represent(image_path, model_name="VGG-Face", enforce_detection=False)
			facial_area = result[0]['facial_area']
			
			# Draw rectangle and text
			cv2.rectangle(img, (facial_area['x'], facial_area['y']), (facial_area['x'] + facial_area['w'], facial_area['y'] + facial_area['h']), (0, 255, 0), 2)
			cv2.putText(img, file, (facial_area['x'], facial_area['y'] - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)
			
			# Save to temp_result folder
			output_path = os.path.join(output_dir, file)
			cv2.imwrite(output_path, img)
```
