```python
import torch
import numpy as np
import os
```

```python
model = torch.hub.load('harritaylor/torchvggish', 'vggish')
model.eval()
```

```python
wav_training_root = 'datasets/extracted/audioData/trainingData/'
output_root = 'datasets/extracted/Features/audio/trainingData/'

for wav_file in os.listdir(wav_training_root):
   file_name = wav_file.split('.wav')[0]
   
   if not os.path.exists(os.path.join(output_root, file_name)):
      os.makedirs(os.path.join(output_root, file_name))
   else:
      if len(os.listdir(os.path.join(output_root, file_name))) > 15:
         print(f"Features for {file_name} already exist, skipping extraction.")
         continue
      
   feature = model.forward(os.path.join(wav_training_root, wav_file))

   feature = feature.cpu().detach().numpy()

   for i in range(feature.shape[0]):
      if not os.path.exists(os.path.join(output_root, file_name)):
         os.makedirs(os.path.join(output_root, file_name))
         
      output = os.path.join(os.path.join(output_root, file_name), f'feature{i}.npy')
      np.save(output, feature[i])
   
   print(f"Features for {file_name} extracted and saved.")
      
print("Feature extraction completed.")



# feature = model.forward('datasets/extracted/audioData/trainingData/_0bg1TLPP-I.004.wav')

# output_folder = 'datasets/extracted/Features/audio/trainingData/_0bg1TLPP-I.004'


# if not os.path.exists(os.path.dirname(output_folder)):
# 	os.makedirs(os.path.dirname(output_folder))

# np.save(output_folder, feature)

# np.save(output_folder, feature.detach().numpy())
```
