```python
dataset_root = "/mnt/d/aa-kuliah/Skripsi/Dataset/first-impressions"
```

```python
train_file_names = os.listdir('datasets/extracted/Features/visual/trainingData')
len(train_file_names)

output_dir = "datasets/extracted/audioData/trainingData"
train_file_names = [f"{f}.mp4" for f in train_file_names]

try:
   if not os.path.exists(output_dir):
      os.makedirs(output_dir)
except OSError:
   print("Error: Creating directory of data")

for file_name in train_file_names:
   video_path = os.path.join(dataset_root, "train", file_name)
   output_path = os.path.join(output_dir, f"{os.path.splitext(file_name)[0]}.wav")
 
   if os.path.exists(output_path):
      print(f"File {file_name}.wav already exists, skipping conversion.")
      continue
   else:
      print(f"Converting {file_name}.mp4 to {file_name}.wav...")
      command = f"ffmpeg -i {video_path} -ab 320k -ac 2 -ar 44100 -vn {output_path}"
      subprocess.call(command, shell=True)
```

```python
val_file_names = os.listdir('datasets/extracted/Features/visual/validationData')
len(val_file_names)

output_dir = "datasets/extracted/audioData/validationData"
val_file_names = [f"{f}.mp4" for f in val_file_names]

try:
   if not os.path.exists(output_dir):
      os.makedirs(output_dir)
except OSError:
   print("Error: Creating directory of data")

for file_name in val_file_names:
   video_path = os.path.join(dataset_root, "val", file_name)
   output_path = os.path.join(output_dir, f"{os.path.splitext(file_name)[0]}.wav")
 
   if os.path.exists(output_path):
      print(f"File {file_name}.wav already exists, skipping conversion.")
      continue
   else:
      print(f"Converting {file_name}.mp4 to {file_name}.wav...")
      command = f"ffmpeg -i {video_path} -ab 320k -ac 2 -ar 44100 -vn {output_path}"
      subprocess.call(command, shell=True)
```
